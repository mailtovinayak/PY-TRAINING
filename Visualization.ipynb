```python
import pandas as pd
wine_review = pd.read_csv("DATA.csv")
```

```python
wine_review
```

# SYNTAX OF MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

# BARCHART OF TOP10 Most Common Wine Varietie

```python
plt.figure(figsize=(12,6))
wine_review["variety"].value_counts().head(10).plot(kind="bar", color="darkred")
plt.title("Top 10 Most Common Wine Varieties")
plt.xlabel("Wine Variety")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# COUNTRY WISE DISTRIBUTION

```python
plt.figure(figsize=(8,8))
wine_review["country"].value_counts().head(5).plot(kind="pie", autopct="%1.1f%%", colormap="coolwarm")
plt.title("Top 5 Countries by Wine Count")
plt.ylabel("")
plt.show()
```

# SCATTER PLOT WINE PRICES VS RATING

```python
plt.figure(figsize=(10,6))
plt.scatter(wine_review["price"], wine_review["points"], alpha=0.5, color="blue") #Points in plot are very close under $200 so most of the 
#  good ratings of wines are affordable under 200 dollars
plt.title("Wine Price vs. Rating")
plt.xlabel("Price ($)")
plt.ylabel("Points")
plt.xlim(0, 500)
plt.grid(True)
plt.show()
```

# box plot price distribution by country

```python
top_countries = wine_review["country"].value_counts().head(5).index
filtered_df = wine_review[wine_review["country"].isin(top_countries)]

plt.figure(figsize=(35,20))
filtered_df.boxplot(column="price", by="country", grid=False)
plt.title("Price Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Price ($)")
plt.ylim(0,200)
plt.show()
```

# average wine price by rating

```python
avg_price_per_point = wine_review.groupby("points")["price"].mean()  ## If to use aggregation function, you have to groupby

plt.figure(figsize=(10,5))
plt.plot(avg_price_per_point.index, avg_price_per_point.values, marker="o", color="green")
plt.title("Average Wine Price by Rating")
plt.xlabel("Wine Points")
plt.ylabel("Average Price ($)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
