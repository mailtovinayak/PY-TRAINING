```python
def extract_year (title):
    year = ""
    for i in title:
        if str.isdigit(i):
            year += i #isdigit is getting numbers
    if year == "":
        year = -1
    else:
        year = int(year)
    
            
    return year
```

```python
wine_review
```

```python
wine_review["year"]=wine_review["title"].apply(extract_year)
```
